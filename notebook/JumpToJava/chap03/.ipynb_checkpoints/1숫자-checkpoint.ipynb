##  숫자 (Number)

### 정수

정수는 양의 정수, 0, 음의 정수를 통틀어서 부르는 말.

자 정수는 크기에 따라서 주로 int, long 타입에 담는다.

````java
int positiveInteger = 2147483647;
int negativeInteger = -2147483648;
````

int 범위를 넘어서는 숫자를 입력할 때는 해당 숫자를 입력하고 뒤에 L 혹은 l 표기를 한 뒤에 long 타입 변수에 담는다.

````java
long positiveInteger = 2147483648L;
long negativeInteger = -2147483649l;
````

빼면 무슨 일이 벌어지냐구?

````java
long positiveInteger = 2147483648;
long negativeInteger = -2147483649;
````

### 실수

정수를 제외한 유리수와 무리수, 즉 점(period)을 사용하는 숫자를 말한다.

유리수는 rational number 로, 두정수의 나눗셈으로 표시할 수 있는 수로, 분수라고도 할 수 있다.

무리수는 irrational number 실수 중에서 유리수가 아닌 수. 즉 분수로 표현할 수 없는 수다. 루트, 파이 등으로 표현을 한다.

````java
float pi = 3.14F;
double morePi = 3.1445648152346854961534894231579816d;
````

자바에서 실수표현은 double 이 기본이라서, double 값 뒤에는 D 나 d 를 입력하지 않아도 된다.

````java
double morePi = 3.1445648152346854961534894231579816;
````

### 8진수와 16진수

8진수와 16진수는 점이 없는 숫자이므로 int 로 표현할 수 있다.

````java
int octal = 023; // 자지별*승수  => 2*8^1 + 3*8^0
int hex = 0xC; // 0...9,A,B,C  => 12
````

###  숫자연산

이것까지 해야하나? 핵심은 코드는 평가되고 그 결과가 치환된다는 것.



다음의 변수가 있다고 가정.

```
int a = 3;
int b = a+3;
```

다음 코드가 어떤 과정을 거치는가?

```
a+b
```

먼저는 앞에서부터 평가(계산)되고. 그 결과가 치환됨

```
3+b
```

다음은 b 가 수식으로 평가되고 그 결과로 치환됨

```
3+a+3
```

a 가 숫자로 평가되고 치환됨

```
3+3+3
```

평가(계산)됨

```
9
```
 

나머지 연산자만 다뤄보자.

````java
return 7 % 3;
````

###  증감연산( ++, --)

위치만 주의하자.

변수는 평가(계산)되고 이후에 치환(substitution) 된다. 

````java
int a = 0;

// ++a 에서 a 라는 변수가 평가되기 전에 1을 더한다.
// 결과는 1이다.
System.out.println(++a); 

int b = 0;
// a++ 에서 a 라는 변수가 평가된 후에 1을 더한다.
// 결과는 0이다.
System.out.println(b++); 

````
