```java
// a 라는 이름을 가진 공간을 만들고
// 정수(int) 만 담을 수 있다고 선언
int a;
```

### 자료형(Type)

변수명 앞에 써있는 게 자료형임. 
자료형은 위 장난감 처럼 모양을 이야기함.

- 정수 모양 변수공간에는 정수만 넣을 수 있음
- 문자열 모양 변수공간에는 문자열만 넣을 수 있음

모양은 형이고 형은 영어로 타입이라고 표현함.

### 변수에 값 대입(할당)

변수 공간에 값을 넣는 것을 할당한다 혹은 대입한다라고 표현함.

할당은 allocation, 대입은 assign.

근데 보통은 할당이라고 한국말로 쓰고 영어로 assign 이라고 표시함.

혼용하는 경향.

```java
int a = 2;
```

```java
int a = "문자열 타입의 값을 입력하면?";
```

### 사용자 정의 자료형

임의의 타입을 만들 수도 있다. 바로 아래처럼.

```java
class 네모라는형{
    public void 짖어(){
        System.out.println("넹몽");
    }
}

네모라는형 네모상자;

네모상자 = new 네모라는형();
네모상자.짖어();
```

네모상자라고 이름이 붙은 변수공간에는 네모라는 형 값만 할당할 수 있다. 

아래처럼 네모상자에는 문자열 형 값은 담을 수 없다.

즉 변수 선언시 주어진 타입과 다른 타입을 가진 값을 할당하면 오류가 난다.

imcompatible type 이라면서.

```java
class 네모라는형{
    public void 짖어(){
        System.out.println("넹몽");
    }
}

네모라는형 네모상자;

네모상자 = new 네모라는형();

네모상자 = "문자열타입의 값 할당";
```
